```python
import tensorflow as tf
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64,activation='relu',input_shape=(10,)), 
    tf.keras.layers.Dense(32,activation='relu'),
    tf.keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=[])
```

```python
sample_data = np.random.rand(100, 10)
```

```python
model.build((None, 10))
```

```python
# Extract weights from a specific layer
weights, biases = model.layers[0].get_weights()

# Plot histogram of weights
plt.hist(weights.flatten(), bins=30, edgecolor='black',color='pink')
plt.title("Weight Histogram for Layer 1")
plt.xlabel("Weight Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# Plot histogram of weights
plt.hist(biases.flatten(), bins=30, edgecolor='black',color='pink')
plt.title("Biases Histogram for Layer 1")
plt.xlabel("Bias Values")
plt.ylabel("Frequency")
plt.show()
```
